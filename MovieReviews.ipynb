## Building Feature Set

**NLTK classifiers require data to be in this format:  
Array of Tuples, where each Tuple has Dictionary of features and category.**

**Splitting training and testing data**

```python
training_documents = documents[0:1500]
testing_documents = documents[1500:]
```

```python
# Creating array containing all the words

all_words = []
for doc in training_documents:
    all_words += doc[0]  #training-documents is an array containing tuples of (word, category)
                         # doc[0] -> words, doc[1] -> category
# all_words
```

```python
# Finding frequency distribution

import nltk
```

```python
freq = nltk.FreqDist(all_words)
freq
```

```python
freq.most_common(15)  # first 15 most frequent words
```

```python
freq.most_common(3000)  # an array of tuples
```

**Features:-**

```python
common = freq.most_common(3000)
features = [i[0] for i in common]  # features contains only words and not frequencies
                                   # Here, features is a list or array containing 3000 words or features
features
```

Now, we have to create a dictionary for each document, stating True or False as value -> wheather each feature comes in each document or not  
Keys - Features  
Values - True or False

```python
training_documents[0]  # tuple of list of words and the category of document. 
                       # Ex: ([word1, word2, word3,...], category of doc.)
```

```python
# Creating Dictionary

def get_feature_dictionary(words):  # words -> list of words in tuple ([word1, word2, ...], category)
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set  # w -> key,  value -> True - if w is in document
                                              #                     False - if w not in document
    return current_features
```

```python
get_feature_dictionary(training_documents[0][0])  
```

```python
training_data = [(get_feature_dictionary(doc), category)for doc, category in training_documents]
testing_data = [(get_feature_dictionary(doc), category)for doc, category in testing_documents]
```

```python
training_data[0]
```

### using Naive Bayes Classifier

Classifier needs data to be in the format:-  
list or array of tuples, where each tuple contains dictionary of features and category of document, each dictionary conatins boolean value of each feature, whether that particular feature is present in our document or not.  

Syntax:-  

**[({feature1.1:True/False, feature1.2:True/False}, category1), ({{feature2.1:True/False, feature2.2:True/False}}, category2), ({{feature3.1:True/False, feature3.2:True/False}}, category3), ...]**

```python
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)  # training classifier
```

```python
# nltk has classify module having accuracy() function

nltk.classify.accuracy(classifier, testing_data)  # finding accuracy of this classifier on the testing data
```

-> we've got an accuracy of 0.776 or 77.6%
